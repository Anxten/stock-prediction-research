```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Tentukan simbol (BBRI.JK) dan rentang waktu
symbol = "BBRI.JK"
start_date = "2019-01-01"
end_date = "2026-01-01"

# Download data
print(f"Sedang mendownload data {symbol}...")
df = yf.download(symbol, start=start_date, end=end_date)

# Tampilkan 5 data teratas
df.head()
```

```python
# Pastikan folder data sudah ada
os.makedirs('../data', exist_ok=True)

# Simpan
df.to_csv('../data/BBRI_raw.csv')
print("Data berhasil disimpan di folder data/BBRI_raw.csv")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], color='blue', label='Harga Close')
plt.title(f'Tren Harga Saham {symbol} (2019 - Sekarang)')
plt.xlabel('Tahun')
plt.ylabel('Harga (IDR)')
plt.legend()
plt.grid(True)
plt.show()
```
